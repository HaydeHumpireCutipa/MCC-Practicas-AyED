```python
from math import *
import matplotlib.pyplot as plot
from numpy import *
```

```python
def Entrada(nombre):
    with open("./"+nombre,"r") as archivo:
        con=0
        x=[]
        y=[]
        for linea in archivo:
            if con==0:
                num_puntos=int(linea)
            else:
                aux=linea.split()
                aux2=[]
                x.append(int(aux[0]))
                for i in range(1,len(aux)):
                    aux2.append(float(aux[i]))
                y.append(aux2)
            con=con+1
    return num_puntos,x,y
```

**Python**

```python
Y=[]
Desv=[]
archivos=["InsertionPy.txt","CountingPy.txt","HeapPy.txt","QuickPy.txt"]

for i in archivos:
    num,x,y=Entrada(i)
    aux=[]
    aux2=[]
    for i in y:
        aux.append(mean(i))
        aux2.append(std(i))
    Y.append(aux)
    Desv.append(aux2)
```

```python
algoritmos=len(archivos)
```

```python
fig, ax = plot.subplots(nrows=1, ncols=1, figsize=(10,7.5))

colors=["b","g","r","c","m","y","k","greenyellow","lime","orange"]

for k in range(algoritmos):
    ax.plot(x,Y[k], linestyle='--', marker='o', color=colors[k],label=archivos[k]) 
ax.legend()

ax.set_xlabel('Numero de Elementos')
ax.set_ylabel('Tiempo (micro-seg)')

# Establecer título
plot.suptitle("Grafica Nro Datos VS Tiempo")

plot.grid(True)

# Mostrar
plot.show()
```

Limites

```python
k=0
aux1=[]
aux2=[]

for i in range(len(Y[k])):
    aux1.append(Y[k][i]+2*Desv[k][i])
    aux2.append(Y[k][i]-2*Desv[k][i])

fig, ax = plot.subplots(nrows=1, ncols=1, figsize=(10,7.5))

colors=["b","g","r","c","m","y","k","greenyellow","lime","orange"]

ax.plot(x,Y[k], linestyle='--', marker='o', color=colors[0],label=archivos[k]) 
ax.plot(x,aux1, linestyle='--', marker='o', color=colors[1],label="Limite Superior")
ax.plot(x,aux2, linestyle='--', marker='o', color=colors[2],label="Limite Inferior")

ax.legend()

ax.set_xlabel('Numero de Elementos')
ax.set_ylabel('Tiempo (micro-seg)')

# Establecer título
plot.suptitle("Grafica Nro Datos VS Tiempo")

plot.grid(True)

# Mostrar
plot.show()
```
