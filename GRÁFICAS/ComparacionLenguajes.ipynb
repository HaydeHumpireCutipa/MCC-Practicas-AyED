```python
from math import *
import matplotlib.pyplot as plot
from numpy import *
```

```python
def Entrada(nombre):
    with open("./"+nombre,"r") as archivo:
        con=0
        x=[]
        y=[]
        for linea in archivo:
            if con==0:
                num_puntos=int(linea)
            else:
                aux=linea.split()
                aux2=[]
                x.append(int(aux[0]))
                for i in range(1,len(aux)):
                    aux2.append(float(aux[i]))
                y.append(aux2)
            con=con+1
    return num_puntos,x,y
```

**Counting Sort en Golang / C++ / Python**

```python
Y=[]
Desv=[]
archivos=["CountingGo.txt","CountingC.txt","CountingPy.txt"]

for i in archivos:
    num,x,y=Entrada(i)
    aux=[]
    aux2=[]
    for i in y:
        aux.append(mean(i))
        aux2.append(std(i))
    Y.append(aux)
    Desv.append(aux2)
```

```python
algoritmos=len(archivos)
fig, ax = plot.subplots(nrows=1, ncols=1, figsize=(10,7.5))

colors=["b","g","r","c","m","y","k","greenyellow","lime","orange"]

for k in range(algoritmos):
    ax.plot(x,Y[k], linestyle='--', marker='o', color=colors[k],label=archivos[k]) 
ax.legend()

ax.set_xlabel('Numero de Elementos')
ax.set_ylabel('Tiempo (micro-seg)')

# Establecer t√≠tulo
plot.suptitle("Grafica Nro Datos VS Tiempo")

plot.grid(True)

# Mostrar
plot.show()
```
